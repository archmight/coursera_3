# обратные ссылки в регулярных выражениях

попробуем найти слова, в которых есть 2 одинаковые идущие буквы

```python
match = re.findall(r"(\w)\1", i_love)
print(match)
```

```python
match = re.findall(r"(\b\w*(\w)\2\w*\b)", i_love)
print(match)
```

```python
match = [_[0] for _ in match]
print(match)
```

```python
match = re.findall(r"(\b\w*(\w)(\w)\2\3\w*\b)", i_love)
print(match)
```

```python
match = re.findall(r"(\b\w*(\w{2})\2\w*\b)", i_love)
print(match)
```
